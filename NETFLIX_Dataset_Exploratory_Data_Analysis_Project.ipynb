categorize Netflix shows into custom popularity levels like "Popular", "Average", "Not Popular"

```python
# Map existing rating categories to popularity levels
popularity_map = {
    'TV-14': 'Popular',
    'PG-13': 'Popular',
    'TV-PG': 'Popular',
    'R': 'Average',
    'TV-MA': 'Average',
    'PG': 'Popular',
    'TV-Y7': 'Not Popular',
    'TV-Y7-FV': 'Not Popular',
    'G': 'Not Popular',
    'TV-G': 'Not Popular',
    'NC-17': 'Average',
    'NR': 'Not Popular',
    'Unknown': 'Not Popular'
}

# Apply the mapping
df['popularity'] = df['rating'].map(popularity_map).fillna('Not Popular')

# Display result
print(df[['title', 'rating', 'popularity']].head(10))

# Check how many shows fall into each category
print("\nCount by popularity level:\n", df['popularity'].value_counts())

# (Optional) Sort by popularity level
popularity_order = ['Popular', 'Average', 'Not Popular']
df_sorted = df.sort_values(by='popularity', key=lambda col: col.map({v: i for i, v in enumerate(popularity_order)}))

# Preview
print("\nSorted by popularity:\n", df_sorted[['title', 'rating', 'popularity']].head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: assuming 'data' contains your top 10 countries
# e.g. data = df_cleaned['country'].value_counts().head(10)

# Create a color palette with as many colors as bars
colors = sns.color_palette("Set2", len(data))  # Try "husl", "tab10", "pastel", etc.

# Plot the bar chart with custom colors
plt.figure(figsize=(10,6))
sns.barplot(x=data.index, y=data.values, palette=colors)

# Customize labels and title
plt.xticks(rotation=90)
plt.ylabel("Count of Contents")
plt.xlabel("Country")
plt.title(" Top 10 Countries with Highest Number of Contents on Netflix", fontsize=13, weight='bold')

plt.tight_layout()
plt.show()
```

## x) Top 10 Genres with the Most Content on Netflix

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combine all genres into one list
all_genres = ', '.join(df['listed_in']).split(', ')
unique_genres = list(set(all_genres))

# Count occurrences of each genre
genre_counts = {genre: all_genres.count(genre) for genre in unique_genres}

# Get the top 10 genres
top_genres = dict(sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)[:10])

# Create a color palette with different colors for each bar
colors = sns.color_palette("husl", len(top_genres))  # You can try "Set2", "tab10", "pastel", etc.

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x=list(top_genres.values()), 
    y=list(top_genres.keys()), 
    palette=colors
)

# Add labels and title
plt.xlabel('Number of Contents', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.title('Top 10 Genres with the Most Content on Netflix', fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```
